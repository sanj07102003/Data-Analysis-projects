```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def draw_plot():
    # 1. Import data
    df = pd.read_csv("https://raw.githubusercontent.com/freeCodeCamp/boilerplate-sea-level-predictor/main/epa-sea-level.csv")

    # 2. Scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])

    # 3. Line of best fit (ALL data) using numpy
    slope, intercept = np.polyfit(df['Year'], df['CSIRO Adjusted Sea Level'], 1)
    years_extended = np.arange(df['Year'].min(), 2051)
    line_all = slope * years_extended + intercept
    plt.plot(years_extended, line_all, color='red', label='Fit: 1880-present')

    # 4. Line of best fit (from year 2000)
    df_recent = df[df['Year'] >= 2000]
    slope_recent, intercept_recent = np.polyfit(
        df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'], 1
    )
    years_recent = np.arange(2000, 2051)
    line_recent = slope_recent * years_recent + intercept_recent
    plt.plot(years_recent, line_recent, color='green', label='Fit: 2000-present')

    # 5. Labels and title
    plt.xlabel("Year")
    plt.ylabel("Sea Level (inches)")
    plt.title("Rise in Sea Level")
    plt.legend()

    # 6. Save and return
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```
